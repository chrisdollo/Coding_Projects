```python
# used to import text files or .csv files into python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# # tensorflow can be used to create Neural Networks
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.losses import SparseCategoricalCrossentropy
```

```python
# import the training data from the cvs file
path_to_file = "/Users/chrisdollo/Desktop/Coding_Projects/digit-recognizer/train.csv"
training_data = pd.read_csv(path_to_file)

# # .head() prints the first five lines of a dataset
# print(training_data.head())

training_data = training_data.to_numpy()

# import the test data from the cvs file
path_to_file_2 = "/Users/chrisdollo/Desktop/Coding_Projects/digit-recognizer/test.csv"
test_data = pd.read_csv(path_to_file_2)
test_data = test_data.to_numpy()


# .shape gives the shape of the data set
# print(training_data.shape)

# separate the training  data between the input and the target
cutOff = int((len(training_data)*80)/100)

X_train = training_data[:cutOff, 1:]
Y_train = training_data[:cutOff, 0]

X_test = training_data[cutOff:, 1:]
Y_test = training_data[cutOff:, 0]


print("The shape of the training data is", X_train.shape)
print("The shape of the training data is", Y_train.shape)

print("The shape of the test data is", X_test.shape)
print("The shape of the test data is", Y_test.shape)
```

```python
# Let's visualize the data
test_image = X_train[12345]
test_image = test_image.reshape((28, 28))

plt.imshow(test_image, cmap='gray')  # cmap='gray' for grayscale images
plt.colorbar()  # Show a colorbar to understand the pixel values
plt.show()
```

```python
# Training the model
model = Sequential([
    Dense(units=25, activation='relu'),
    Dense(units=15, activation='relu'),
    Dense(units=10, activation='linear')
])

model.compile(loss=SparseCategoricalCrossentropy(from_logits=True))

model.fit(X_train, Y_train, epochs=100)
```

```python
# function that evaluates the model
def evaluate_model(prediction, target):
    """
    evaluate the error of a model

    Parameters:
    - prediction (int): array containing predictions from the model.
    - target (int): array containing actual output.

    Returns:
    - a percentage of error
    """
    
    if (len(prediction) != len(target)):
        return -1
    
    numberOfExamples = len(prediction)
    wrong = 0

    for i in range(len(prediction)):
        if prediction[i] != target[i]:
            wrong+=1
        
    return ((wrong/numberOfExamples)*100)
```

```python
def getPredictions(output):
    """
    Select the highest probability label for every image

    Parameters:
    - output (int array): 
    Returns:
    - a prediction class for each training example in the set
    """
    predictions = list()
    
    for arrayOfProbabilities in output:
        maxPrediction , maxPredictionIndex = 0, 0
        for i in range(len(arrayOfProbabilities)):
            if arrayOfProbabilities[i] > maxPrediction:
                maxPrediction = arrayOfProbabilities[i]
                maxPredictionIndex = i
        predictions.append(maxPredictionIndex)
        
    return predictions
```

```python
# Evaluation of the model on the training data
output = model(X_train)
f_x = tf.nn.softmax(output)
modelPrediction = getPredictions(f_x)
errorTrainingSet = evaluate_model(modelPrediction, Y_train)

# # Evaluation of the model on the test data
# output = model(X_test)
# f_x = tf.nn.softmax(output)
# modelPrediction = getPredictions(f_x)
# errorTestSet = evaluate_model(modelPrediction, Y_test)


print("The error on the training set is: ",errorTrainingSet)
# print("The error on the testing set is: ",errorTestSet)
```

```python
# convert array into dataframe and export the cvs file
column_labels = ['Label'] 
DF = pd.DataFrame(predictions, columns=column_labels)
DF.index = DF.index + 1
DF.to_csv("data1.csv", index=True, index_label='ImageID')
```
